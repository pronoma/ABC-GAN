### Import Friedman1 dataset


Friedman 1 Dataset 

*   $y(X) = 10 * sin(pi * X_0 * X_1) + 20 * (X_2 - 0.5) ** 2 + 10 * X_3 + 5 * X_4 + noise * N(0, 1)$.
* Only 5 features used to calculate y 
* Noise is Gaussian 
* 1000 datapoints and 10 features used in the following experiment 


```python
n_features = 10 
X, Y = friedman1Dataset.friedman1_data(n_samples,n_features)
```

## Stats Model

```python
[coeff,Ypred] = statsModel.statsModel(X,Y)
```

**Observation**: the coefficients of the stats model from x6 to x10 are close to zero. Since only 5 features out of 10 are used to calculate y in the actual model 

# GAN Model

```python
lr = 0.01 
beta_1 = 0.5 
beta_2 = 0.999
```

```python
real_dataset = dataset.CustomDataset(X,Y)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
generator = network.ComplexGenerator(coeff).to(device)
discriminator = network.Discriminator(coeff,25).to(device)
criterion = torch.nn.BCEWithLogitsLoss()
gen_opt = torch.optim.Adam(generator.parameters(), lr=lr, betas=(beta_1, beta_2))
disc_opt = torch.optim.Adam(discriminator.parameters(), lr=lr, betas=(beta_1, beta_2))
```

```python
n_epochs = 30000
batch_size = n_samples//2
```

```python
train_test.training_GAN(discriminator,generator,disc_opt,gen_opt,real_dataset,batch_size, n_epochs,criterion,device)
```

```python
train_test.test_generator(generator,real_dataset,device)
```

# ABC GAN Model

### ABC - GAN Model 

```python
#Initialize Generator and Discriminator 
disc = network.Discriminator(coeff,100).to(device)
gen = network.ComplexGenerator(coeff).to(device)

criterion = torch.nn.BCEWithLogitsLoss()
gen_opt = torch.optim.Adam(gen.parameters(), lr=lr, betas=(beta_1, beta_2))
disc_opt = torch.optim.Adam(disc.parameters(), lr=lr, betas=(beta_1, beta_2))
```

```python
n_epochs = 10000
batch_size = n_samples//2
```

```python
ABC_train_test.training_GAN(disc,gen,disc_opt,gen_opt,real_dataset,n_epochs,batch_size,criterion,coeff,mean,variance,device)
```

```python
#Test the generator 
ABC_train_test.test_generator(gen,real_dataset,coeff,mean,variance,device)
```

**Sanity Check**

```python
sanityChecks.discProbVsError(real_dataset,disc,device)
```

**Visualisation of Trained GAN Generator**

```python
Layer12 =[]
Layer23 =[]
Output =[]
for name,param in gen.named_parameters():
  if name == "hidden1.weight":
    Layer12 = np.transpose(np.absolute(param.data.detach().cpu().numpy()))
  if name == "hidden2.weight":
    Layer23 = np.transpose(np.absolute(param.data.detach().cpu().numpy()))
  if name == "output.weight":
    Output = np.transpose(np.absolute(param.data.detach().cpu().numpy()))
```

```python
sns.set(rc = {'figure.figsize':(25,12)})
ax = sns.heatmap(Layer12)
```

```python
sns.set(rc = {'figure.figsize':(50,12)})
mulL1L2 = Layer12.dot(Layer23)
ax = sns.heatmap(mulL1L2)
```

```python
sns.set(rc = {'figure.figsize':(1,12)})
mulL2L3 =mulL1L2.dot(Output)
ax = sns.heatmap(mulL2L3)
```
