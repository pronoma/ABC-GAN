```python
import pandas as pd
import numpy as np
import math
from statistics import mean, stdev
from sklearn import preprocessing
import pandas as pd
```

```python
#Dataset Size
n = 5000
variance = 1 
```

```python
x = np.random.uniform(0,1,(n,5))
df_x = pd.DataFrame(x, columns = ['X_0','X_1','X_2','X_3','X_4'])
print(df_x)
# Scale X 
scaler = preprocessing.StandardScaler()
df = pd.DataFrame(scaler.fit_transform(df_x), columns=df_x.columns)
print(df.head())
X_scaled = df.iloc[:,0:n]
X_scaled = X_scaled.to_numpy()
```

```python
y_tmp = 10 * np.sin(math.pi * X_scaled[:,0]* X_scaled[:,1]) + 20 * np.square(X_scaled[:,2]+0.5) + 10 * X_scaled[:,3] + 5* X_scaled[:,4]
mu = mean(y_tmp)
std = stdev(y_tmp)
y_true = (y_tmp - mu)/ std 
y = y_true + np.random.normal(0,variance,y_true.shape)
y = np.reshape(y,(n,1))
```

```python
dataset = np.concatenate((x,y), axis = 1)
df = pd.DataFrame(dataset,columns=['X0','X1','X2','X3','X4','Y'])
display(df)
```

```python
df.to_csv("quadratic_constant_variance_dataset.csv")
```
